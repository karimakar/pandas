```python
# let create dataframe
data = pd.DataFrame(np.random.randn(10000,4))
data.head()
```

```python
data.shape
```

```python
# descriptive statistic
data.describe()
```

```python
# chose one column
col0 = data[0]
col0
```

```python
# detect oulier base on the rule of 3 std that mean each value greather or less than 3 std is considered like outlier
print(col0[np.abs(col0>3)].count())
col0[np.abs(col0>3)]
```

```python
col1 = data[1]
print(col1[np.abs(col1>3)].count())
col1[np.abs(col1>3)]
```

```python
col2 = data[2]
print(col2[np.abs(col2>3)].count())
col2[np.abs(col2>3)]
```

```python
col3 = data[3]
print(col3[np.abs(col3>3)].count())
col3[np.abs(col3>3)]
```

```python
sns.displot(data[0],kde=True)
sns.displot(data[1],kde=True)
sns.displot(data[2],kde=True)
sns.displot(data[3],kde=True)
```
