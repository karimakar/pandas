# read our data

```python
import pandas_datareader.data as web
import pandas as pd
```

```python
all_data = {ticker:web.get_data_yahoo(ticker)
           for ticker in ['VLO','OXY','F','GOOG']}
```

```python
price = pd.DataFrame({ticker:data['Adj Close']
                     for ticker,data in all_data.items()})
```

```python
volume = pd.DataFrame({ticker:data["Volume"]
                      for ticker ,data in all_data.items()})
```

```python
price
```

```python
volume
```

```python
returns = price.pct_change()
```

```python
returns
```

```python
returns.tail()
```

```python
returns['VLO'].corr(returns['F'])
```

```python
returns['OXY'].corr(returns['F'])
```

```python
returns['VLO'].cov(returns['F'])
```

```python
returns['GOOG'].cov(returns['F'])
```

```python
returns.VLO.cov(returns.F)
```

```python
returns.corr()
```

```python
returns.cov()
```

```python
returns.corrwith(returns.F)
```

```python
returns.corrwith(volume)
```

```python
returns.GOOG.corrwith(returns.F)
```
